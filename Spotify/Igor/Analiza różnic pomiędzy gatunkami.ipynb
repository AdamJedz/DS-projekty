```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
#matplotlib inline
# ^^^ pyforest auto-imports - don't write above this line
df = pd.read_csv('../train_set.csv')
df = df.rename(columns={"Unnamed: 0": "index"}).set_index("index")
df.drop(['track_id', 'track_album_id', 'playlist_id'], axis = 1, inplace=True)
```

### Rozkłady zmiennych dla poszczególnych gatunków

```python
g = sns.FacetGrid(df, col = 'playlist_genre', hue = 'playlist_genre', col_wrap = 3)
g.map(sns.distplot, 'track_popularity', kde = False)
plt.subplots_adjust(top = .9)
g.fig.suptitle('Rozkład zmiennej track_popularity')
```

Widzimy, że dla każdego z gatunków występuje nadreprezentacja zerowej popularności. 
Niemniej, praktycznie wszystkie rozkłady wyglądają bardzo podobnie. 
Rzeczami, które moga zastanawiać jest szczyt wykresu dla gatunku *edm*, który jest przesunięty na lewo względem pozostałych oraz gwałtowne ucięcie wykresu w okolicach 80 dla *rocka*.

```python
g = sns.FacetGrid(df, col = 'playlist_genre', hue = 'playlist_genre', col_wrap = 3)
g.map(sns.distplot, 'danceability', kde = False)
plt.subplots_adjust(top = .9)
g.fig.suptitle('Rozkład zmiennej danceability')
```

Rozkłady wszystkich gatównków, oprócz *rocka* są lewostronnie skośne. Rozkład *rocka* jest bardzo zbliżony do normalnego.

```python
g = sns.FacetGrid(df, col = 'playlist_genre', hue = 'playlist_genre', col_wrap = 3)
g.map(sns.distplot, 'energy', kde = False)
plt.subplots_adjust(top = .9)
g.fig.suptitle('Rozkład zmiennej energy')
```
