```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
#matplotlib inline
# ^^^ pyforest auto-imports - don't write above this line
df = pd.read_csv('../train_set.csv')
df = df.rename(columns={"Unnamed: 0": "index"}).set_index("index")
df.drop(['track_id', 'track_album_id', 'playlist_id'], axis = 1, inplace=True)
```

### Rozkłady zmiennych dla poszczególnych gatunków

```python
g = sns.FacetGrid(df, col = 'playlist_genre', hue = 'playlist_genre', col_wrap = 3)
g.map(sns.distplot, 'track_popularity', kde = False)
plt.subplots_adjust(top = .9)
g.fig.suptitle('Rozkład zmiennej track_popularity')
```

Widzimy, że dla każdego z gatunków występuje nadreprezentacja zerowej popularności. 
Niemniej, praktycznie wszystkie rozkłady wyglądają bardzo podobnie. 
Rzeczami, które moga zastanawiać jest szczyt wykresu dla gatunku *edm*, który jest przesunięty na lewo względem pozostałych oraz gwałtowne ucięcie wykresu w okolicach 80 dla *rocka*.

```python
g = sns.FacetGrid(df, col = 'playlist_genre', hue = 'playlist_genre', col_wrap = 3)
g.map(sns.distplot, 'danceability', kde = False)
plt.subplots_adjust(top = .9)
g.fig.suptitle('Rozkład zmiennej danceability')
```

Rozkłady wszystkich gatównków, oprócz *rocka* są lewostronnie skośne. Rozkład *rocka* jest bardzo zbliżony do normalnego.

```python
g = sns.FacetGrid(df, col = 'playlist_genre', hue = 'playlist_genre', col_wrap = 3)
g.map(sns.distplot, 'energy', kde = False)
plt.subplots_adjust(top = .9)
g.fig.suptitle('Rozkład zmiennej energy')
```

Statystycznie najmniej do tańczenia nadają się *r&b* i *rap*, a najbardziej kawałki *edm* i *rockowe*.

```python
g = sns.FacetGrid(df, col = 'playlist_genre', hue = 'playlist_genre', col_wrap = 3)
g.map(sns.distplot, 'loudness', kde = False)
plt.subplots_adjust(top = .9)
g.fig.suptitle('Rozkład zmiennej loudness')
```

Bardzo ciekawe jest to, że najwięcej skrajnych przypadkó dla zmiennej loudness jest dla gatunku *latin*. 
To może oznaczać, że ujemne wartości są najgłośniejsze, a te bliższe zeru najcichsze. To jest kolejna rzecz do zbadania.

```python
g = sns.FacetGrid(df, col = 'playlist_genre', hue = 'playlist_genre', col_wrap = 3)
g.map(sns.countplot, 'mode')
plt.subplots_adjust(top = .9)
g.fig.suptitle('Rozkład zmiennej loudness')
```

Wszędzie przeważa major jako typ skali. Najmniejsza różnica jest dla *r&b*, a największa dla *rocka*.

```python
g = sns.FacetGrid(df, col = 'playlist_genre', hue = 'playlist_genre', col_wrap = 3)
g.map(sns.distplot, 'speechiness', kde = False)
plt.subplots_adjust(top = .9)
g.fig.suptitle('Rozkład zmiennej speechiness')
```

Jak można było się spodziewać, mowy w utworach jest bardzo mało. Jedynym wyjątkiem jest *rap*.

```python
g = sns.FacetGrid(df, col = 'playlist_genre', hue = 'playlist_genre', col_wrap = 3)
g.map(sns.distplot, 'acousticness', kde = False)
plt.subplots_adjust(top = .9)
g.fig.suptitle('Rozkład zmiennej acousticness')
```

Zdecydowana mniejszość kawałków jest akustyczna. Najbardziej to widać w *edm*. Ckiut większą ilośc utworów akustycznych widać w *r&b*, *latin* i *rapie*.

```python
g = sns.FacetGrid(df, col = 'playlist_genre', hue = 'playlist_genre', col_wrap = 3)
g.map(sns.distplot, 'instrumentalness', kde = False, bins = 10)
plt.subplots_adjust(top = .9)
g.fig.suptitle('Rozkład zmiennej instrumentalness')
```

Wiemy, że ta zmienna przyjmuje wartość 1 gdy w piosence nie ma żadnych słów. Dlatego też nic dziwnego, że najwięcej piosenek z ta zmienną różną od zera jest w gatunku *edm*.

```python
g = sns.FacetGrid(df, col = 'playlist_genre', hue = 'playlist_genre', col_wrap = 3)
g.map(sns.distplot, 'liveness', kde = False, bins = 15)
plt.subplots_adjust(top = .9)
g.fig.suptitle('Rozkład zmiennej liveness')
```

Jeżeli chodzi o rozkład występów na żywo, to w każdym gatunku wygląda to podobnie.

```python
g = sns.FacetGrid(df, col = 'playlist_genre', hue = 'playlist_genre', col_wrap = 3)
g.map(sns.distplot, 'valence', kde = False, bins = 15)
plt.subplots_adjust(top = .9)
g.fig.suptitle('Rozkład zmiennej valence')
```

Rozkład zmiennej valence pokazuje bardzo ciekawą zależność:  
* *Edm* i *rap* częściej nastrajają negatywnie (zwłaszcza *edm*),
* *Latin* i *r&b* pozytywnie (zwłaszcza *latin*),
* *Rock* i *pop* najczęściej są odbierane neutralnie.

```python
g = sns.FacetGrid(df, col = 'playlist_genre', hue = 'playlist_genre', col_wrap = 3)
g.map(sns.distplot, 'tempo', kde = False, bins = 25)
plt.subplots_adjust(top = .9)
g.fig.suptitle('Rozkład zmiennej tempo')
```

Ciekawostką jest to, ze dla gatunku *edm*. Ponad połowa utworów ma bardzo podobne tempo (ok 120 bpm).

```python
g = sns.FacetGrid(df, col = 'playlist_genre', hue = 'playlist_genre', col_wrap = 3)
g.map(sns.distplot, 'duration_ms', kde = False, bins = 25)
plt.subplots_adjust(top = .9)
g.fig.suptitle('Rozkład zmiennej duration_ms')
```

Rozkłady długości trwania wyglądają podobnie dla wszystkich gatunków. Jedyną różnicą jest to, że część *rockowych* i *edm* trwa dłużej od reszty.
